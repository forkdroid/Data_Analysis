EXPLORATORY DATA ANALYSIS

Gender

```python
ax = sns.countplot(x='Gender', data =df)
for bars in ax.containers:
  ax.bar_label(bars)
```

We observe from the above graph that most sales were done by females

Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

Men & Women in the age group 26-35 did the maximum shopping

State


```python
state = df.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state, x = 'State',y= 'Orders', palette='Set2')
```

Maximum sales took place in the state of Uttar Pradesh

Occupation


```python
ax = sns.countplot(data=df, x = 'Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

Most buyers were from the IT, HealthCare sector

Product Category

```python

ax = sns.countplot (data = df , x = 'Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

The product category purchased the most was Clothing and Apparel
